```python
def to_df(beams):
    jsons = [beam.dict() for beam in beams]
    return pd.DataFrame(sorted(jsons, key=lambda json: json['total_log_likelihood'], reverse=True))
```

4つ放射管において

4, つの, 放射, 管 としたいが、冒頭のマッチングが

4, つ放, ...     になってしまうと嬉しくない

4, つ, の, 放射   になって欲しい

AC を ABCを前から比較していく
* A (distance: 1)), C (distance: 0) とみなすか
* AC (distance: 1)
とみなすか2通りあるという話

今回の場合 "つの" とマッチするような撮り方は
* つの -> つ, 
* つの → つ放
があり、よりマッチする後者の方を選択した結果ごちゃごちゃになってしまった。

しかし、"挿あ入" を"挿入"に直すとき
* 挿 がマッチングする
* あ入 と　入 を前から比較すると、distance=1の候補が2つ出てくる
  * あ (-> '')
  * あ入 (-> '入')


短く取っていってdistanceをちゃんと再計算すれば問題なさそう。
