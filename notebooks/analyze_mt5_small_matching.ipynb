```python
from spell_correction.t5_spell_corrector import T5SpellCorrector
```

```python
import pandas as pd
```

```python
spell_corrector = T5SpellCorrector()
```

```python
def to_df(beams):
    jsons = [beam.dict() for beam in beams]
    return pd.DataFrame(sorted(jsons, key=lambda json: json['total_log_likelihood'], reverse=True))
```

* "彼はファヴローにクラッシック西部劇映画を提供した。", # 'クラッシック' → 'クラシック'
* "民主党の予備選挙ではヒラリー・クリントンが1位となり、バラク・オバマが僅差の2位だた。",  # 'だた' → 'だった'
* "クラゲには4つ放射管を持つが触手はもたない。",  # '' -> 'の'
* "桶狭間の戦いで国友制の鉄砲が織田信長により初めて戦力として使用された。",  # '制' → '製'
* "ひきこもりの更正をはじめたきっかけ",  # '正' → '生'

```python
beams = spell_corrector.beam_search("彼はファヴローに <extra_id_0>西部劇映画を提供した。", original_token="クラッシック")
to_df(beams)
```

```python
beams = spell_corrector.beam_search("民主党の予備選挙ではヒラリー・クリントンが1位となり、バラク・オバマが僅差の2位 <extra_id_0>。", original_token="だた")
to_df(beams)
```

```python
beams = spell_corrector.beam_search("桶狭間の戦いで <extra_id_0>の鉄砲が織田信長により初めて戦力として使用された。", original_token="国友制")
to_df(beams)
```
