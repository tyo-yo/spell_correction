```python
from spell_correction.t5_spell_corrector import T5SpellCorrector
import pandas as pd
```

```python
spell_corrector = T5SpellCorrector(
    model_name="google/mt5-small",
    filter_func=lambda c: c["distance"] <= 1 and c["match_len"] >= 1,
    add_original_token_to_masked_text=False  # つけたい時は自分でつけるので
)
```

```python
## 実験内容
```

```python
masked_text_0 = "桶狭間の戦いで <extra_id_0>の鉄砲が織田信長により初めて戦力として使用された。"
original_token = '国友制'

candidates_0 = spell_corrector.calc_next_token_candidates(masked_text_0, original_token)
pd.DataFrame(candidates_0[:10])
```
